```python
time_convert = udf(time_convertor, IntegerType())
df = df.withColumn("cookTime",time_convert(df.cookTime))
```

```python
df = df.withColumn('ingredients', lower(col('ingredients'))) \
       .filter(df.ingredients.like("%beef%")) \
       .withColumn("prepTime", regexp_replace(col("prepTime"), "[^0-9]", "")) \
       .withColumn("cookTime", regexp_replace(col("cookTime"), "[^0-9]", "")) \
       .withColumn('total_cook_time',reduce(add, [col(x) for x in ['prepTime','cookTime']])) \
       .withColumn(
                   'difficulty',
                    when((col("total_cook_time") <= 30), 'easy')\
                   .when((col("total_cook_time") <= 60), 'medium')\
                   .when((col("total_cook_time") > 60), 'hard')\
                   .otherwise(0)
                  )
```

```python
df_avg = df.groupby("difficulty").agg(avg(col("total_cook_time")))
df_avg.show()
```
